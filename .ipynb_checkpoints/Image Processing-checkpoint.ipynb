```python
def compute_disparity(im0, im1, loss, W):
    image_h, image_w = im0.shape
    vertical_strtix = 0
    vertical_endix = W
    disparity = np.zeros_like(im0)
    while vertical_strtix < image_h:
        compare_strtix = 0
        compare_endix = W
        while compare_strtix < image_w:
            disp_arr = []
            center_arr = []
            horizontal_strtix = compare_strtix
            horizontal_endix = compare_endix
            cur_center = (horizontal_strtix+horizontal_endix)//2
            while horizontal_strtix < image_w:
                disp_arr.append(loss(im0[vertical_strtix:vertical_endix, compare_strtix:compare_endix], im1[vertical_strtix:vertical_endix, horizontal_strtix:horizontal_endix]))
                center_arr.append((horizontal_strtix+horizontal_endix)//2)
                horizontal_strtix = horizontal_endix
                horizontal_endix+=1
            disp = np.array(disp_arr).argmax()
            d = center_arr[disp] - cur_center
            #disparity[vertical_strtix:vertical_endix, compare_strtix:compare_endix] = d
            disparity[(vertical_strtix+vertical_endix)//2, cur_center] = d
            compare_strtix = compare_endix
            compare_endix+=1
        vertical_strtix = vertical_endix
        vertical_endix+=1
    return disparity                
```

```python
def sq_loss(mat1, mat2):
    r, c = mat2.shape
    return np.sum(np.square(mat1[:r, :c]-mat2))
```

```python
def ab_loss(mat1, mat2):
    r, c = mat2.shape
    return np.sum(np.abs(mat1[:r, :c] - mat2))
```

```python
result = compute_disparity(im1,im0, ab_loss, 7)
```
